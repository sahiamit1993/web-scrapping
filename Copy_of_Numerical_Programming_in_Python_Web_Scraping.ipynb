## **Scrapping Movie Title**

````python
# Write Your Code here
def scrape_movie_info(movie_link):
    response = requests.get(movie_link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Example: Modify based on the actual HTML structure
        title_element = soup.find('div', class_='title-block').find('h1')


        # Check if the elements were found before accessing the text attribute
        title = title_element.text.strip() if title_element else 'N/A'

        # Print or store the extracted information
        print(f"{title}")


# Loop through each movie link and scrape information
for link in movie_links:
    scrape_movie_info(link)
````

````python
# Write Your Code here
def scrape_movie_info(movie_link):
    response = requests.get(movie_link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Example: Modify based on the actual HTML structure
        release_date_element = soup.find('span', class_='text-muted')

        # Check if the elements were found before accessing the text attribute
        release_date = release_date_element.text.strip() if release_date_element else 'N/A'

        # Print or store the extracted information
        print(f"{release_date}")


# Loop through each movie link and scrape information
for link in movie_links:
    scrape_movie_info(link)
````

## **Scrapping Genres**

````python
# Write Your Code here
def scrape_movie_info(movie_link):
    response = requests.get(movie_link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the element with class "detail-infos__subheading" containing "Age rating"
        genres = soup.find('h3', class_='detail-infos__subheading', string='Genres')

# If the element is found, get the corresponding value
        if genres:
           genres = genres.find_next('div', class_='detail-infos__value').text.strip()
           print(genres)
        else:
           print("not found")
# Loop through each movie link and scrape information
for link in movie_links:
    scrape_movie_info(link)
````

````python
def scrape_movie_info(movie_link):
    response = requests.get(movie_link)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        rating_span = soup.find('span', text=re.compile(r'\d+\.\d+'))
        if rating_span:
           rating = rating_span.text.strip().split()[0]
           print(f"{rating}")
        else:
           print("No rating found in the HTML string.")
for link in movie_links:
    scrape_movie_info(link)
````

## **Analyzing Top Genres**

````python
# prompt: Analyzing Top Genres

# **Movies**
# Get the top 5 most frequent genres
top_genres = df['Genres'].apply(lambda x: x.split(','))
top_genres = np.concatenate(top_genres)
top_genres = pd.Series(top_genres).value_counts().head(5)

# Print the top 5 genres and their counts
print("Top 5 Genres for Movies:")
print(top_genres)

# **TV Shows**
# Get the top 5 most frequent genres
top_genres_tv = df_tv['Genres'].apply(lambda x: x.split(','))
top_genres_tv = np.concatenate(top_genres_tv)
top_genres_tv = pd.Series(top_genres_tv).value_counts().head(5)

# Print the top 5 genres and their counts
print("\nTop 5 Genres for TV Shows:")
print(top_genres_tv)
````

````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a word cloud for movie genres
movie_genres = ' '.join(top_genres.index.to_list())
wordcloud_movie = WordCloud(width=800, height=400).generate(movie_genres)

# Create a word cloud for TV show genres
tv_genres = ' '.join(top_genres_tv.index.to_list())
wordcloud_tv = WordCloud(width=800, height=400).generate(tv_genres)

# Plot the word clouds
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.imshow(wordcloud_movie, interpolation='bilinear')
plt.axis("off")
plt.title("Movie Genres")

plt.subplot(1, 2, 2)
plt.imshow(wordcloud_tv, interpolation='bilinear')
plt.axis("off")
plt.title("TV Show Genres")

plt.show()
````
